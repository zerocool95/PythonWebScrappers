## Put product page reviews link in "amazon_rev_url"

```python
amazon_rev_url = "https://www.amazon.in/Apple-iPhone-X-64GB-Silver/product-reviews/B0711T2L8K/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1"#+str(page_num)
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'
print(amazon_rev_url)
headers = {'User-Agent': user_agent}
page = requests.get(amazon_rev_url, headers = headers)
parser = html.fromstring(page.content)
xpath_reviews = '//div[@data-hook="review"]'
reviews = parser.xpath(xpath_reviews)
xpath_rating  = './/i[@data-hook="review-star-rating"]//text()' 
xpath_title   = './/a[@data-hook="review-title"]//text()'
xpath_author  = './/a[@data-hook="review-author"]//text()'
xpath_date    = './/span[@data-hook="review-date"]//text()'
xpath_body    = './/span[@data-hook="review-body"]//text()'
# xpath_helpful = './/span[@data-hook="helpful-vote-statement"]//text()'
xpath_total_reviews = './/span[@data-hook="total-review-count"]//text()'
xpath_verified = './/span[@data-hook="avp-badge"]//text()'
xpath_helpful_votes = './/span[@data-hook="helpful-vote-statement"]//text()'
```

```python
total_reviews = int(parser.xpath(xpath_total_reviews)[0].replace(",",""))
total_pages = total_reviews//10
print("Total_Pages: ", total_pages)
```

```python
cnt = 0
reviews_df = pd.DataFrame(columns = ['rating', 'title', 'author', 'date', 'body', 'verified','helpful_votes', 'sentiment'])
for page_num in range(1, total_pages):
    amazon_rev_url = "https://www.amazon.in/Apple-iPhone-X-64GB-Silver/product-reviews/B0711T2L8K/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber="+str(page_num)
    user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'
    print(amazon_rev_url)
    headers = {'User-Agent': user_agent}
    page = requests.get(amazon_rev_url, headers = headers)
    parser = html.fromstring(page.content)
    xpath_reviews = '//div[@data-hook="review"]'
    reviews = parser.xpath(xpath_reviews)
    xpath_rating  = './/i[@data-hook="review-star-rating"]//text()' 
    xpath_title   = './/a[@data-hook="review-title"]//text()'
    xpath_author  = './/a[@data-hook="review-author"]//text()'
    xpath_date    = './/span[@data-hook="review-date"]//text()'
    xpath_body    = './/span[@data-hook="review-body"]//text()'
    # xpath_helpful = './/span[@data-hook="helpful-vote-statement"]//text()'
    xpath_total_reviews = './/span[@data-hook="total-review-count"]//text()'
    xpath_verified = './/span[@data-hook="avp-badge"]//text()'
    xpath_helpful_votes = './/span[@data-hook="helpful-vote-statement"]//text()'
    for i, review in enumerate(reviews):
        rating  = float(review.xpath(xpath_rating)[0].split()[0].replace(",",""))
        title   = review.xpath(xpath_title)[0]
        author  = review.xpath(xpath_author)
        date    = review.xpath(xpath_date)[0]
        body    = review.xpath(xpath_body)[0]
#         helpful = review.xpath(xpath_helpful)[0]
        if len(review.xpath(xpath_helpful_votes)) > 0:
            helpful_votes = review.xpath(xpath_helpful_votes)[0].split()[0].replace(",","")
        if helpful_votes == 'One':
            helpful_votes = 1
        helpful_votes = int(helpful_votes)
        if len(review.xpath(xpath_verified)) > 0:
            verified = review.xpath(xpath_verified)[0]
            if verified == 'Verified Purchase':
                verified = True
            else:
                verified = False
        else:
            verified = False
            
        if rating < 3: 
            sentiment = 'Negative'
        else:
            sentiment = 'Positive'
        
        review_dict = {'rating': rating,
                       'title': title,
                       'author': author,             
                       'date': date,
                       'body': body,
                        "verified" : verified,
                        "helpful_votes" : helpful_votes,
                        "sentiment" : sentiment}
        print("\n----------{}---------\n".format(cnt), end = '\r')
        cnt += 1
#         pprint(review_dict)
        reviews_df = reviews_df.append(review_dict, ignore_index=True)
    
        
        
```

```python
def normalize(df):
    result = df.copy()
    feature_name = 'rating'
    
    max_value = df[feature_name].max()
    min_value = df[feature_name].min()
    print(min_value,max_value)
    result['sentiment_range'] = 2*((df[feature_name] - min_value) / (max_value - min_value)) - 1.0
    return result
```

```python
reviews_df = normalize(reviews_df)
```
