```python
import string
from lxml import html
from json import dump,loads
from requests import get
import json
from re import sub
from dateutil import parser as dateparser
from time import sleep
from lxml import html  
import requests
from pprint import pprint
import pandas as pd
from tqdm import tqdm
```

```python
base_url = 'https://www.medicinenet.com/script/main/alphaidx.asp?p=a_dict'

user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'
```

```python
headers = {'User-Agent': user_agent}
page = requests.get(base_url, headers = headers)
parser = html.fromstring(page.content)
```

```python
xpath_list = "//div[@class='AZ_results']/ul/li/a/text()"
```

```python
reviews = parser.xpath(xpath_list)
```

```python
reviews[::-1]
```

```python
med_terms = []
for c in tqdm(string.ascii_lowercase, total = 26):
    base_url = 'https://www.medicinenet.com/script/main/alphaidx.asp?p='+str(c)+'_dict'
    page = requests.get(base_url, headers = headers)
    parser = html.fromstring(page.content)
    med_terms += parser.xpath(xpath_list)
```

## Post processing on Med_terms extracted from website

```python
new_med_terms = []
for term in med_terms:
    if len(term.strip()) > 3:
        new_med_terms.append(term.lower())
        
```

```python
import pickle
with open('med_terms.pkl', 'wb') as f:
    pickle.dump(new_med_terms, f)
```

```python
with open('med_terms.pkl', 'rb') as f:
    gg = pickle.load(f)
```
